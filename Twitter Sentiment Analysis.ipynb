# Preprocessing Data

```python
# removes pattern in the input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
# remove twitter handles (@user)
train0['clean_text'] = np.vectorize(remove_pattern)(train0['tweet'], "@[\w]*")
```

```python
# for test
# remove twitter handles (@user)
test0['clean_text'] = np.vectorize(remove_pattern)(test0['tweet'], "@[\w]*")
```

```python

def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = text.replace("[^a-zA-Z#]", " ")
    
    return text
```

```python
train0['clean_text'] = train0['clean_text'].apply(lambda x:clean_text(x))
train0.head()
```

```python
#for test
test0['clean_text'] = test0['clean_text'].apply(lambda x:clean_text(x))
test0.head()
```

# Modeling

### Input Split

```python
#Vectorizing
# feature extraction
from sklearn.feature_extraction.text import CountVectorizer

#bag of words

bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000)
bow = bow_vectorizer.fit_transform(train2['clean_text'])
```

```python
#vectorizing test
bow_test = bow_vectorizer.fit_transform(test2['clean_text'])
```

```python
#bow[0].toarray()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bow, train2['label'], random_state=23, test_size=0.25)
```

### Model Training

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
```

```python
# training
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```python
# testing
pred = LR.predict(X_test)
f1_score(y_test, pred)
```

```python
accuracy_score(y_test, pred)
```

```python
# use probability to get output
pred_prob = LR.predict_proba(X_test)
pred = pred_prob[:, 1] >= 0.3
pred1 = pred.astype(np.int)

f1_score(y_test, pred)
```

```python
accuracy_score(y_test,pred)
```

Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
# testing
pred = NB_classifier.predict(X_test)
print(f1_score(y_test, pred))
print(accuracy_score(y_test,pred))
```
